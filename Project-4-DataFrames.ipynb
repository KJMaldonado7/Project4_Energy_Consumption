```python
import pandas as pd
```

```python
# Read in CSV file
all_energy_df = pd.read_csv("Resources/recs2020_public_v7.csv")
all_energy_df.head()
```

```python
# Create DataFrame with building characteristics
housing_characteristics_df = all_energy_df[['DOEID', 'TYPEHUQ', 'YEARMADERANGE', 'TOTROOMS', 'WALLTYPE', 'ROOFTYPE', 'ADQINSUL', 'NUMFRIG', 'EQUIPM', 
                                            'ACEQUIPM_PUB', 'TOTSQFT_EN', 'TOTALBTU', 'TOTALDOL']].copy()
housing_characteristics_df.set_index('DOEID',inplace=True)
housing_characteristics_df.head()
```

```python
# Create DataFrame with information on energy-efficient upgrades
energy_efficient_df = all_energy_df[['DOEID','DWCYCLE', 'TVTYPE1', 'TYPETHERM', 'LGTINLED', 'SMARTMETER', 'SOLAR', 'ELECVEH', 
                                     'TOTALBTU', 'TOTALDOL']].copy()
energy_efficient_df.set_index('DOEID',inplace=True)
energy_efficient_df.head()
```

```python
# Create DataFrame with regional information
regional_df = all_energy_df[['DOEID','REGIONC', 'DIVISION', 'state_postal', 'BA_climate', 'TOTALBTU', 'TOTALDOL']].copy()
regional_df.set_index('DOEID',inplace=True)
regional_df.head()
```

```python
# Create DataFrame with demographic information
demographic_df = all_energy_df [['DOEID', 'HHSEX', 'HHAGE', 'EMPLOYHH', 'EDUCATION', 'SDESCENT', 'HOUSEHOLDER_RACE', 
                                'NHSLDMEM', 'NUMCHILD', 'MONEYPY', 'TOTALBTU', 'TOTALDOL']].copy()
demographic_df.set_index('DOEID',inplace=True)
demographic_df.head()
```

```python
# Export them all as CSVs
housing_characteristics_df.to_csv('Table_CSVs/housing_characteristics.csv')
energy_efficient_df.to_csv('Table_CSVs/energy_efficient.csv')
regional_df.to_csv('Table_CSVs/regional.csv')
demographic_df.to_csv('Table_CSVs/demographics.csv')
```
